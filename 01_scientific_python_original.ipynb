**Quicksort**

Implement the quicksort algorithm, as defined by wikipedia

```python
from cProfile import label
from dataclasses import dataclass
import random
def partition(A, L, H):                    # A =  an array of values
                                           # L = Lower Position
                                           # H = Higher Position
    
    index = L                              # initilize the starting index
                        
    pivot = A[index]                       #  Choose the first element as the pivot
    while L < H:
        while L < len(A) and A[L] <= pivot: # Increment (element larger than the pivot)
            L = L + 1
        while A[H] > pivot:                 # Decrement (element smaller than pivot)
            H = H - 1
        if L < H:                           # swap the number on lower and higher position
            A[L], A[H] = A[H], A[L]    
            
    A[H], A[index] = A[index], A[H] 
    return H                               #Returning end pointer to divide the array into 2



def quicksort(A, L, H):
    if len(A) < 2:
        return "To perfom quicksort, at least two elements are required"
    elif L < H:
        pivotPos = partition(A, L, H)
        quicksort(A, L, pivotPos-1)     #before pivotPos
        quicksort(A, pivotPos+1, H)     #after pivotPos
    return A



if __name__ == '__main__':
    random.seed(10)
    A = random.sample(range(1, 100), 12)
    print('Unsorted List is:',  A)
    print('Sorted List is: %s' % quicksort(A ,0 , len(A)-1))
    
```

**1.6.5.1 Curve fitting**

The temperature extremes in Alaska for each month, starting in January, are given by (in degrees Celcius):

max:  17,  19,  21,  28,  33,  38, 37,  37,  31,  23,  19,  18

min: -62, -59, -56, -46, -32, -18, -9, -13, -25, -46, -52, -58


1. Plot these temperature extremes.


```python
# Create a array of min and max temperature and define the 12 months for 1 year
temp_max = np.array([17,  19,  21,  28,  33,  38, 37,  37,  31,  23,  19,  18])
temp_min = np.array([-62, -59, -56, -46, -32, -18, -9, -13, -25, -46, -52, -58])
Months = np.arange(12)


# Plotting the temperature
fig, ax = plt.subplots()
plt.plot(Months,temp_max,  marker='o', color="blue", linewidth=2.0, linestyle="-")
plt.plot(Months,temp_min,  marker='o', color="red", linewidth=2.0, linestyle="-")
plt.gca().legend(('Max Temp','Min Temp'))
ax.set_ylabel('Temperature')
ax.set_xlabel('Month')
```

2. Define a function that can describe min and max temperatures. 

Hint: this function has to have a period of 1 year. 
Hint: include a time offset.

```python
# the periodic function look like y=AcosB(x+c)+D
# A = Mean - peak
# c = constant = 0
# D = mean  
# B = 2*pi/period

def temp_fun(T, mean, amplitude, T_off):
    return (amplitude * np.cos((T + T_off) * 2 * np.pi / T.max())+ mean)
```

3. Fit this function to the data with scipy.optimize.curve_fit().

4. Plot the result. Is the fit reasonable? If not, why?



```python
params_max, cov_max = optimize.curve_fit(temp_fun, Months, temp_max, [20, 10, 0])
params_min, cov_min = optimize.curve_fit(temp_fun, Months, temp_min, [-40, 20, 0])

#plot the result

plt.figure()
plt.plot(Months,temp_max,  marker='o', color="blue", linewidth=2.0, linestyle="-")
plt.plot(Months,temp_min,  marker='o', color="red", linewidth=2.0, linestyle="-")
plt.plot(Months, temp_fun(Months, *params_max), 'r--')
plt.plot(Months, temp_fun(Months, *params_min), 'b--')
plt.gca().legend(('Max Temp Data','Min Temp Data','Max Temp Fun','Min Temp Fun'))
plt.xlabel('Month')
plt.ylabel('Temperature')
plt.show()

```


From the above result, we can say that the function follows the response of the data. However, we need to find if the response is within the the acceptable accuracy or not. For that we need to calculate the fit accuracy. We will use  Mean Absolute Percentage Error (MAPE). 

```python
temp_max_pred = temp_fun(Months, *params_max)
temp_min_pred = temp_fun(Months, *params_min)

def MAPE(y_true,y_pred):
    mape = np.mean(np.abs((y_true - y_pred)/y_true))*100
    return mape

print('Maxmimum Temperature Curve Fitting Error:',MAPE(temp_max,temp_max_pred))
print('Minimum Temperature Curve Fitting Error:',MAPE(temp_min,temp_min_pred))
```

From the above analysis,  we found that MAPE for temp_max curve is 3.1006778138476756%, while MAPE for temp_min curve is 8.629593873542284%. We could say the fit is resonable for the max temp curve while fit is not resonable for min temperature curve as the MAPE is high.

**1.6.5.2. Finding the minimum of a scalar function**


**2-D minimization**


The six-hump camelback function

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def camelback(x):
    return ((4 - 2.1*x[0]**2 + x[0]**4 / 3.) * x[0]**2 + x[0] * x[1]
            + (-4 + 4*x[1]**2) * x[1] **2)
    
# Make a grid to evaluate the function (for plotting)
x = np.linspace(-2, 2)
y = np.linspace(-1, 1)

xg, yg = np.meshgrid(x, y)
plt.figure()
plt.imshow(camelback([xg, yg]), extent=[-2, 2, -1, 1], origin="lower")
plt.colorbar()
res = optimize.minimize(camelback, x0=[0, 0])
plt.scatter(res.x[0], res.x[1])
plt.show()
```

There is one global minima, and the function value is [0,0].

**1.6.8. Fast Fourier transforms: scipy.fftpack**

**Denoise moon landing image**


1. Examine the provided image moonlanding.png, which is heavily contaminated with periodic noise. In this exercise, we aim to clean up the noise using the Fast Fourier Transform.
2. Load the image using matplotlib.pyplot.imread().


```python
import numpy as np 
import matplotlib.pyplot as plt 
from scipy import fftpack
from skimage.color import rgb2gray
from skimage import io

img_array=plt.imread("moonlanding.png") 
img_array_gray = rgb2gray(img_array)
plt.figure(num=None, figsize=(8, 6), dpi=80)
plt.imshow(img_array, cmap = 'gray')
plt.title('Moonlanding Original Gray Image') 
plt.axis('off')
plt.show()
```

3. Find and use the 2-D FFT function in scipy.fftpack, and plot the spectrum (Fourier transform of) the image. Do you have any trouble visualising the spectrum? If so, why?

```python
## Fourier transformation of the gray image
img_fft = fftpack.fft2(img_array_gray) 
plt.figure(num=None, figsize=(8, 6), dpi=80)
plt.imshow(np.log(abs(img_fft)), cmap = 'gray')
plt.title('Fourier Transform') 
plt.show()
```

4. The spectrum consists of high and low frequency components. The noise is contained in the high-frequency part of the spectrum, so set some of those components to zero (use array slicing).

```python
#Remove all the high frequencies (array sclicing)
keep_fraction = 0.1
img_hf_fft = img_fft.copy()
row, col = img_hf_fft.shape
#set zero for row and column inside the fraction range 
img_hf_fft[int(row*keep_fraction):int(row*(1-keep_fraction))] = 0
img_hf_fft[:,int(col*keep_fraction):int(col*(1-keep_fraction))] = 0
plt.figure(num=None, figsize=(8, 6), dpi=80)
plt.imshow(np.log(abs(img_hf_fft)), cmap = 'gray')
plt.title('Filtered Image') 
plt.show()
```

5. Apply the inverse Fourier transform to see the resulting image.

```python
#Reconstruct the final Image (inverse fourier transform)
img_recon = fftpack.ifft2(img_hf_fft).real
plt.figure(num=None, figsize=(8, 6), dpi=80)
plt.imshow(np.log(abs(img_recon)), cmap = 'gray')
plt.title('Reconstructed Image') 
plt.show()
```

```python
ntimes = 1

results = []
for n in [2, 10] + list(range(500, 8000, 500)):

    print(f'\n------------ n is {n}')
    
    a = np.random.uniform(size=(n, n)).astype(np.float32)
    b = np.random.uniform(size=(n, n)).astype(np.float32)
    print('np type', a.dtype)

    at = torch.from_numpy(a)
    bt = torch.from_numpy(b)
    print('torch type', at.dtype)

    atgpu = at.to('cuda')
    btgpu = bt.to('cuda')
    print('cuda type', atgpu.dtype)
    
    start_time = time.time()
    for i in range(ntimes):
        c = a @ b
    nptime = time.time() - start_time
    print(f'Numpy took {nptime:.3f} seconds. {c[0, 0:5]}')

    start_time = time.time()
    for i in range(ntimes):
        ct = at @ bt
    torchtime = time.time() - start_time
    print(f'Pytorch took {torchtime:.3f} seconds. {ct[0, 0:5]}')

    start_time = time.time()
    for i in range(ntimes):
        ctgpu = atgpu @ btgpu
    cudatime = time.time() - start_time
    print(f'CUDA took {torchtime:.3f} seconds. {ct[0, 0:5]}')

    results.append([n, nptime, torchtime, cudatime])
```

```python
results = np.array(results)

plt.figure(figsize=(12, 12))
plt.plot(results[:, 0], results[:, 1:])
plt.legend(('Numpy', 'Torch', 'Cuda'))
plt.xlabel('$n$')
plt.ylabel(r'Seconds to multiply two $n \times n$ matrices');
```
